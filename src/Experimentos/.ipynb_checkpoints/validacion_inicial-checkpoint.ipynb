```python
import json
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file = open("Piloto/secuencia_facil.json", "r") 
```

```python
json1_str = file.read()
json1_data = json.loads(json1_str)
```

```python
json1_data.keys()
```

```python
del json1_data["start_time"]
del json1_data["stop"]
del json1_data["next"]
```

```python
df = pd.DataFrame(json1_data)
```

```python
len(json1_data["posX"])
```

```python
df.count()
```

Graficamos todos los puntos obtenidos en una prueba para verificar que se logran captar las figuras que realiza el usuario sobre la pizarra.

```python
plt.figure(figsize=(6,9))

for i in range(len(df.posX )-200):
    posX = df.loc[i,"posX"]
    posY = df.loc[i,"posY"]*-1
    
    plt.plot(posX, posY, "o", c="b", markersize=7, alpha=0.5)
plt.xlim([0,900])
plt.ylim([-1900,0])
```

```python
plt.figure(figsize=(15,5))
plt.plot(df["time"].iloc[100:-100], df["velocity"].iloc[100: - 100])
plt.ylabel("Velocity (px/ms)")
plt.xlabel("Time (ms)")
```

```python
plt.figure(figsize=(15,5))
plt.plot(df["time"].iloc[100:-100], df["pressure"].iloc[100: - 100])
plt.ylabel("Pressure")
plt.xlabel("Time (ms)")
plt.ylim([0.15, 0.2])
```
