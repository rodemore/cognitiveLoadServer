```python
import json
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file = open("Piloto/secuencia_facil_2.json", "r") 
```

```python
json1_str = file.read()
json1_data = json.loads(json1_str)
```

```python
json1_data.keys()
```

```python
del json1_data["start_time"]
del json1_data["stop"]
del json1_data["next"]
```

```python
df = pd.DataFrame(json1_data)
```

```python
len(json1_data["posX"])
```

```python
df["time"] = df["time"].apply(lambda t: t - df["time"][100])
```

```python
df  = df[df["time"]<=1000*120]
```

```python
df.tail()
```

```python
df.to_csv("Piloto/secuencia_facil_2.tsv",sep="\t" ,index=False)
```

Graficamos todos los puntos obtenidos en una prueba para verificar que se logran captar las figuras que realiza el usuario sobre la pizarra.

```python
plt.figure(figsize=(6,9))

for i in range(len(df.posX )-200):
    posX = df.loc[i,"posX"]
    posY = df.loc[i,"posY"]*-1
    
    plt.plot(posX, posY, "o", c="b", markersize=7, alpha=0.5)
plt.xlim([0,900])
plt.ylim([-1900,0])
```
