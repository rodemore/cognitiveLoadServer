```python
import json
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file = open("testing2.json", "r") 
```

```python
tasks = open("tasks_result.json", "r") 
```

```python
json1_str = file.read()
json1_data = json.loads(json1_str)
```

```python
json1_data.keys()
```

```python
data = json1_data.copy()
del data["start_time"]
del data["stop"]
del data["next"]
```

```python
df = pd.DataFrame(data)
```

```python
df.head()
```

Graficamos todos los puntos obtenidos en una prueba para verificar que se logran captar las figuras que realiza el usuario sobre la pizarra.

```python
plt.figure(figsize=(6,9))

for i in range(len(df.posX )):
    posX = df.loc[i,"posX"]
    posY = df.loc[i,"posY"]*-1
    
    plt.plot(posX, posY, "o", c="b", markersize=7, alpha=0.5)
plt.xlim([0,1000])
plt.ylim([-2000,0])
plt.savefig("movimiento_recuperado.jpg")
```

```python
plt.figure(figsize=(15,5))
plt.plot(df["time"].iloc[100:-100], df["velocity"].iloc[100: - 100])
plt.ylabel("Velocity (px/ms)")
plt.xlabel("Time (ms)")
```

```python
plt.figure(figsize=(15,5))
plt.plot(df["time"].iloc[100:-100], df["pressure"].iloc[100: - 100])
plt.ylabel("Pressure")
plt.xlabel("Time (ms)")
plt.ylim([0.15, 0.2])
```
